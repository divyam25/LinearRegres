```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
from math import exp
%matplotlib inline
```

```python
a=pd.DataFrame(load_iris().data,load_iris().target)
a = shuffle(a,random_state = 3)
a
```

```python
X_train = a[[0,1,2,3]].head(125).as_matrix()
Y_train = pd.Series(a.index[:125])
Y_train = Y_train.as_matrix()
X_test = a[[0,1,2,3]].tail(25).as_matrix()
Y_test = pd.Series(a.index[125:])
Y_test = Y_test.as_matrix()

Y_train
```

```python
a_train = Y_train.copy()
for i in range(125) :
    if(a_train[i] == 0) :
        a_train[i] = 1
    elif (a_train[i] == 1 or a_train[i] == 2) :
        a_train[i] = 0
a_train
```

```python
b_train = Y_train.copy()
for i in range(125) :
    if (b_train[i] == 2) :
        b_train[i] = 0
b_train
```

```python
c_train = Y_train.copy()
for i in range(125) :
    if(c_train[i] == 2) :
        c_train[i] = 1
    elif (c_train[i] == 1) :
        c_train[i] = 0
c_train
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15,2))
ax[0].plot(Y_train)
ax[0].set_ylabel('Y Train')
ax[1].plot(Y_test)
ax[1].set_ylabel('Y Test');
```

```python
#So far So good... The data seems equally spaced!
theta1 = np.random.RandomState(3).rand(4,1)
theta2 = np.random.RandomState(3).rand(4,1)
theta3 = np.random.RandomState(55).rand(4,1)
theta3
```

```python
for i in range(271) :
    check = 0
    for i in range(125) :
        check = check + ((1/(1+exp(-1 * (theta1.T * np.matrix(X_train[i]).T))))-a_train[i])
    print (check)
    t1 = 0
    for i in range(125) :
        t1 = t1 + ((1/(1+exp(-1 * (theta1.T * np.matrix(X_train[i]).T))))-a_train[i]) * X_train[i][0]
    t2 = 0
    for i in range(125) :
        t2 = t2 + ((1/(1+exp(-1 * (theta1.T * np.matrix(X_train[i]).T))))-a_train[i]) * X_train[i][1]
    t3 = 0
    for i in range(125) :
        t3 = t3 + ((1/(1+exp(-1 * (theta1.T * np.matrix(X_train[i]).T))))-a_train[i]) * X_train[i][2]
    t4 = 0
    for i in range(125) :
        t4 = t4 + ((1/(1+exp(-1 * (theta1.T * np.matrix(X_train[i]).T))))-a_train[i]) * X_train[i][3]

    theta1[0] = theta1[0] - 0.000009 * t1
    theta1[1] = theta1[1] - 0.000009 * t2
    theta1[2] = theta1[2] - 0.000009 * t3
    theta1[3] = theta1[3] - 0.000009 * t4
theta1
```

```python
for i in range(469) :
    check = 0
    for i in range(125) :
        check = check + ((1/(1+exp(-1 * (theta2.T * np.matrix(X_train[i]).T))))-b_train[i])
    print (check)
    t1 = 0
    for i in range(125) :
        t1 = t1 + ((1/(1+exp(-1 * (theta2.T * np.matrix(X_train[i]).T))))-b_train[i]) * X_train[i][0]
    t2 = 0
    for i in range(125) :
        t2 = t2 + ((1/(1+exp(-1 * (theta2.T * np.matrix(X_train[i]).T))))-b_train[i]) * X_train[i][1]
    t3 = 0
    for i in range(125) :
        t3 = t3 + ((1/(1+exp(-1 * (theta2.T * np.matrix(X_train[i]).T))))-b_train[i]) * X_train[i][2]
    t4 = 0
    for i in range(125) :
        t4 = t4 + ((1/(1+exp(-1 * (theta2.T * np.matrix(X_train[i]).T))))-b_train[i]) * X_train[i][3]

    theta2[0] = theta2[0] - 0.0000601 * t1
    theta2[1] = theta2[1] - 0.0000601 * t2
    theta2[2] = theta2[2] - 0.0000601 * t3
    theta2[3] = theta2[3] - 0.0000601 * t4
theta2
```

```python
for i in range(7000) :
    check = 0
    for i in range(125) :
        check = check + ((1/(1+exp(-1 * (theta3.T * np.matrix(X_train[i]).T))))-c_train[i])
    print (check)
    t1 = 0
    for i in range(125) :
        t1 = t1 + ((1/(1+exp(-1 * (theta3.T * np.matrix(X_train[i]).T))))-c_train[i]) * X_train[i][0]
    t2 = 0
    for i in range(125) :
        t2 = t2 + ((1/(1+exp(-1 * (theta3.T * np.matrix(X_train[i]).T))))-c_train[i]) * X_train[i][1]
    t3 = 0
    for i in range(125) :
        t3 = t3 + ((1/(1+exp(-1 * (theta3.T * np.matrix(X_train[i]).T))))-c_train[i]) * X_train[i][2]
    t4 = 0
    for i in range(125) :
        t4 = t4 + ((1/(1+exp(-1 * (theta3.T * np.matrix(X_train[i]).T))))-c_train[i]) * X_train[i][3]

    theta3[0] = theta3[0] - 0.00053 * t1
    theta3[1] = theta3[1] - 0.00053 * t2
    theta3[2] = theta3[2] - 0.00053 * t3
    theta3[3] = theta3[3] - 0.00053 * t4
theta3
```

```python
pred = np.zeros((25,1))
```

```python
for i in range(25) :
    pred1 = (1/(1+exp(-1 * (theta1.T * np.matrix(X_test[i]).T))))
    pred2 = (1/(1+exp(-1 * (theta2.T * np.matrix(X_test[i]).T))))
    pred3 = (1/(1+exp(-1 * (theta3.T * np.matrix(X_test[i]).T))))
    if(pred1>pred2 and pred1>pred3) :
        pred[i] = 0
    if(pred2>pred1 and pred2>pred3) :
        pred[i] = 1
    if(pred3>pred2 and pred3>pred1) :
        pred[i] = 2
```
